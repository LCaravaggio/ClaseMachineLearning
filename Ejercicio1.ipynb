Carga de Librerías


```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

Carga y visualización de datos

```python
df = pd.read_csv('https://raw.githubusercontent.com/LCaravaggio/ClaseMachineLearning/master/alturas-pesos.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(df['Peso'], df['Altura'], alpha=0.5)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['Genero'])
df['Genero_enc'] = le.transform(df['Genero'])

plt.figure(figsize=(14, 7))
plt.scatter(df['Peso'], df['Altura'], c=df['Genero_enc'], alpha=0.5)

plt.plot(c='b')
```

Partición de Dataset

```python
from sklearn.model_selection import train_test_split
X=['Peso', 'Altura']
y=['Genero']
X_train, X_test, y_train, y_test = train_test_split(df[X], df[y], test_size=0.2)
```
